```python
def trans_func(x, xshift = 0.0, xscale = 1.0, maxval = 1.0, minval = 0.0):
    y = -tanh((x-xshift)/xscale)
    ymin, ymax = y.min(), y.max()
    print ymin, ymax
    y = (y - y.min()) * ((maxval - minval) / (ymax - ymin)) + minval
    return y
```

```python
def dens_f(r, z, sig_f, scalhei_f):
    h = scalhei_f(r)
    return 1.0/sqrt(2.0*pi*h) * sig_f(r) * exp(-0.5*(z/h)**2)

def sig_f_Andr(r, Sig_c, r_c, gam):
    t = r / r_c
    s = exp(-gam*log(t))
    return Sig_c * s * exp(-s*t*t)

def sig_f_Ele(r, Sig_c, r_c, gam):
    return ones_like(r)

def scalhei_f(r, h_c, r_c, psi):
    t = r / r_c
    return h_c * exp((psi+1.0) * log(t))

def scale_height_modif(r, a, b, r0, rs):
    try:
        y = b + 0.5 * (a-b) * (1.0 - tanh((r-r0)/rs))
        fdep = 1.0/(y*y) - 1.0
    except ZeroDivisionError:
        print 'Dividing by zero'
        print r, y
        raise
    scal_modi = sqrt(fdep)
    return scal_modi

def dens_f_modi(r, z, sig_f, scalhei_f, scalhei_modi_f, eps=1e-30):
    return dens_f(r, z, sig_f, lambda r: scalhei_f(r)/(scalhei_modi_f(r)+eps))

def ax_imshow(ax, img, x_s, y_s, cmap=None, vmin=None, vmax=None):
    if vmin == None:
        vmin = img.min()
    if vmax == None:
        vmax = img.max()
    if cmap == None:
        cmap = cm.Spectral_r
    ax.imshow(img, origin="lower", vmin=vmin, vmax=vmax,
              cmap=cmap,
              extent=(x_s.min(), x_s.max(), y_s.min(), y_s.max()))
```

```python
Sig_c = 1.0
r_c = 50.0
gam = 1.5
h_c = 10.0
psi = 0.0
```

```python
r_s = linspace(1.0,200.0,num=200)
z_s = linspace(0.0,100.0,num=200)
rr, zz = meshgrid(r_s, z_s)
```

```python
n0 = dens_f(rr, zz,
           lambda r: sig_f_Andr(r, Sig_c, r_c, gam),
           lambda r: scalhei_f(r, h_c, r_c, psi))
```

```python
from mpl_toolkits.axes_grid1 import ImageGrid
```

```python
def make_my_colormap(c_list=None):
    if c_list == None:
        c_list = [(0.0, 'blue'),
                  (0.2, 'CornflowerBlue'),
                  (0.4, 'cyan'),
                  (0.6, 'Chartreuse'),
                  (0.75, 'yellow'),
                  (1.0, 'red')]
    import matplotlib.colors as colors
    return colors.LinearSegmentedColormap.from_list('mycm', c_list)


mycm = make_my_colormap(c_list=[(0.0, (0.5, 0.0, 0.5)),
                                (0.2, (0.0, 0.0, 1.0)),
                                (0.4, (0.0, 0.8, 1.0)),
                                (0.6, (0.0, 0.8, 0.0)),
                                (0.8, (1.0, 0.8, 0.0)),
                                (1.0, (1.0, 0.0, 0.0))])
```

```python
a = 1.0
b = [0.2, 0.5, 0.8]
r0 = [10.0, 40.0, 70.0]
rs = 10.0

import itertools

fig = figure(figsize=(12,12))
grid = ImageGrid(fig, 111, nrows_ncols=(3,3), axes_pad=(0.1,0.15))

i_ax = 0

for b_i,r0_i in itertools.product(b, r0):
    #print b_i, r0_i
    n = dens_f_modi(rr, zz,
                    lambda r: sig_f_Ele(r, Sig_c, r_c, gam),
                    lambda r: scalhei_f(r, h_c, r_c, psi),
                    lambda r: scale_height_modif(r, a, b_i, r0_i, rs),
                   )
    ax = grid[i_ax]
    if i_ax != len(grid)-1:
        ax_imshow(ax, log10(n+1e-30), r_s, z_s, cmap=mycm)
        ax.text(10, 85, '$({0:.1f},\\ {1:.1f})$'.format(b_i, r0_i),
                color="black", fontsize=16)
    
    ax.set_autoscale_on(False)
    ax.set_xlabel("r (AU)")
    ax.set_ylabel("z (AU)")
    i_ax += 1
    
ax_imshow(grid[-1], log10(n0+1e-30), r_s, z_s, vmin=-5.0, vmax=1.0, cmap=mycm)
grid[-1].text(10, 85, r'$n_{\rm gas}$', color="black", fontsize=20)

savefig('ele_deple.pdf', bbox_inches='tight')
```
